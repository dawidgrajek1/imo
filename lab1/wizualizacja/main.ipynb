```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Wczytanie danych
cycles = pd.read_csv('../kroA200.tsp_greedyNN.csv', header=None).T
points = pd.read_csv('../kroA200.tsp_list.csv', header=None)

# list of successive points
cycle1 = cycles[0].values
cycle2 = cycles[1].values
```

```python
# visualization of the tsp problem
plt.figure(figsize=(10, 10))
plt.scatter(points[1], points[2])
# plot the cycles
for i in range(len(cycle1)-1):
    plt.plot([points.iloc[cycle1[i], 1], points.iloc[cycle1[i+1], 1]],
             [points.iloc[cycle1[i], 2], points.iloc[cycle1[i+1], 2]], 'r')
    plt.plot([points.iloc[cycle2[i], 1], points.iloc[cycle2[i+1], 1]],
             [points.iloc[cycle2[i], 2], points.iloc[cycle2[i+1], 2]], 'b')


plt.title('KroA200')
plt.show()
```
